```python
num_features = []

for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        num_features.append(col)

cat_features = []
for col in df.columns:
    if df[col].dtype in ['object','str']:
        cat_features.append(col)

print("-"*140)
print("Numerical Features:", num_features)
print("-"*140)
print("Categorical Features:", cat_features)
```

```python
summary = df[cat_features].describe().T
summary["unique_values"] = df[cat_features].apply(
    lambda col: col.unique().tolist()
)
summary["Unknown_Count"] = df[cat_features].apply(
    lambda col: (col == 'Unknown').sum()
)
summary
```

```python
cols = ['Income_Category', 'Education_Level', 'Marital_Status']

unknown_counts = {c: (df[c] == 'Unknown').sum() for c in cols}
unknown_counts
```

```python
df['unknown_count'] = sum((df[c] == 'Unknown').astype(int) for c in cols)

df['unknown_count'].value_counts().sort_index()
```

```python
pd.crosstab(
    df['unknown_count'],
    df['Attrition_Flag'],
    normalize='index'
)
```

```python

df[cols] = df[cols].replace('Unknown', 'Not_Provided')
```

# Dictionary

- **CLIENTNUM** → Unique customer identifier (ID)

- **Attrition_Flag** → Indicates whether the customer is active (Existing Customer) or has churned (Attrited Customer)

- **Customer_Age** → Customer age (in years)

- **Gender** → Customer gender (M or F)

- **Dependent_count** → Number of dependents of the customer

- **Education_Level** → Customer’s education level

- **Marital_Status** → Customer’s marital status

- **Income_Category** → Customer’s annual income range

- **Card_Category** → Customer’s credit card category (Blue, Silver, Gold, Platinum)

- **Months_on_book** → Length of the customer relationship with the bank (in months)

- **Total_Relationship_Count** → Total number of products the customer has with the bank

- **Months_Inactive_12_mon** → Number of inactive months in the last 12 months

- **Contacts_Count_12_mon** → Number of contacts between the customer and the bank in the last 12 months

- **Credit_Limit** → Customer’s total credit limit

- **Total_Revolving_Bal** → Total revolving balance on the credit card

- **Avg_Open_To_Buy** → Average available credit for use

- **Total_Trans_Amt** → Total transaction amount in the last 12 months

- **Total_Trans_Ct** → Total number of transactions in the last 12 months

- **Total_Amt_Chng_Q4_Q1** → Percentage change in transaction amount from Q1 to Q4

- **Total_Ct_Chng_Q4_Q1** → Percentage change in transaction count from Q1 to Q4

- **Avg_Utilization_Ratio** → Average credit utilization ratio
